# 2. Group by

```python
#grouping data with avg by column ('Continent') dengan sort by asc by column ('2022 Population')

df.groupby('Continent').mean(numeric_only = True).sort_values(by = "2022 Population", ascending = False)
```

```python
#menampilkan column yang di select by desc -> [df.columns[5:13][::-1]]

df2 = df.groupby('Continent')[df.columns[5:13][::-1]].mean(numeric_only = True).sort_values(by = "2022 Population", ascending = False)
df2
```

```python
#transpose column to index from df2 

df3 = df2.transpose()
df3
```

# 3. Visualitation Data

```python
df3.plot()
```

Findings:

Since the 1970s, Asia has dominated the world's population and has continued to grow significantly, especially in the 2000s. However, the growth rate began to stabilize over time. Other continents also experienced population increases, with Oceania showing the lowest growth rate.
